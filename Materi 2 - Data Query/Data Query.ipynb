# Query data dengan logical expression

## Mengambil baris dengan 1 kondisi tertentu

```python
df[df['order_id']=='539991']
```

```python
df.loc[df['order_id']=='539991']
```

```python
df.loc[df['order_id']=='539991', ['product_code','product_name','quantity']]
```

yang sebenarnya terjadi adalah kita membuat 1 binary array yang berisi nilai <mark>False</mark> untuk semua baris kecuali baris yang memenuhi kondisi tersebut diberi nilai <mark>True</mark>, kemudian menjadikannya non-integer index untuk meng-query data

order_id | order_id = '539991'?
--- | ---
539991 | True
10 | False

```python
df['order_id']=='539991'
```

```python
# alternatif penulisan df['order_id']
df.loc[df.order_id=='539991']
```

namun penulisan <mark>df.order_id</mark> (menggunakan titik) tidak dapat digunakan pada kondisi nama kolom mengandung spasi seperti <mark>df.order id</mark>
sehingga penulisan <mark>df.['order id']</mark> perlu digunakan dan sifatnya lebih general (berlaku di semua kondisi, dengan atau tanpa spasi sebagai nama/index kolom)

## Mengambil beberapa baris dengan beberapa kondisi tertentu

### Beberapa kondisi dengan operator & (and)

```python
df[(df['product_name'].str.contains('TEA')) & (df['quantity']==1)]

```

seperti pada 1 kondisi tertentu, pada 2 atau lebih kondisi pun yang sebenarnya terjadi adalah kita membuat binary array untuk setiap kondisi dan mengevaluasinya berdasarkan konsep kombinasi logika

product_name | quantity | product_name mengandung 'TEA'? | quantity = 1? | true/false
--- | --- | --- | --- | ---
TEA | 1 | True | True | True
TEA | 10 | True | False | False
CAKE | 1 | False | True | False
CAKE | 15 | False | False | False

```python
df['product_name'].str.contains('TEA')
```

```python
df['quantity']==1
```

### Beberapa kondisi dengan operator | (or)

```python
df[(df['product_name'].str.contains('TEA')) | (df['product_name'].str.contains('COFFEE'))]
```

product_name | product_name mengandung 'TEA'? | product_name mengandung 'COFFEE'? | true/false
--- | --- | --- | ---
TEA | True | False | True
COFFEE | False | True | True
CAKE | False | False | False

```python
df['product_name'].str.contains('TEA')
```

```python
df['product_name'].str.contains('COFFEE')
```

### Mengambil beberapa baris dengan kombinasi kondisi and ataupun or

```python
df[(df['product_name'].str.contains('TEA') | df['product_name'].str.contains('COFFEE')) & (df['quantity']>10)]
```

product_name | quantity | product_name mengandung 'TEA'? | product_name mengandung 'COFFEE'? | quantity > 10? | true/false

TEA | 5 | True | False | False | False
COFFEE | 50 | False | True | True | True
COFFEE | 7 | False | True | False | False
CAKE | 12 | False | False | True | False
